```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
food = pd.read_excel('./fastfood.xlsx')
```

```python
food
```

```python

plt.bar(food['restaurant'], food['calories'])
plt.xlabel('Restaurante')
plt.ylabel('Calorias')
plt.title('Gráfico de Barras')
plt.show()
```

```python
plt.hist(food['calories'], bins=10, edgecolor='black')
plt.xlabel('Calorias')
plt.ylabel('Valores')
plt.title('Histograma')
plt.show()
```

```python
sns.displot(food['protein'], kde = True)
```

```python
categoria_selecionada = 'Mcdonalds'
dados_filtrados = food.loc[food['restaurant'] == categoria_selecionada]
```

```python
plt.scatter(dados_filtrados['restaurant'], food['calories'])
plt.xlabel('Calorias')
plt.ylabel('Valores')
plt.title('Histograma')
plt.show()
```
